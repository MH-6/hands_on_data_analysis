## 1.5 筛选的逻辑

### 1.5.1 任务一： 我们以"Age"为筛选条件，显示年龄在10岁以下的乘客信息。

```python
df[df['Age']<10].head() # 筛选某列满足某种情况下数据
```

### 1.5.2 任务二： 以"Age"为条件，将年龄在10岁以上和50岁以下的乘客信息显示出来，并将这个数据命名为midage

```python
midage = df[(df['Age']>10) & (df['Age']<50)] # 条件筛选，交集并集
midage.head(10)
```

### 1.5.3 任务三：将midage的数据中第100行的"Pclass"和"Sex"的数据显示出来

```python
midage.loc[[99],['Pclass','Sex']]  #第100行数据的row index number应该是99
```

```python
midage.loc[99,['Pclass','Sex']] # 注意本行代码与上一行代码的区别，row index那里是否用了[]
```

```python
# midage是由原表df筛选得出的，其row index number是不连续的有中断的
# 通过函数reset_index()来重置索引
midage_new = midage.reset_index(drop=True) # 重置索引，返回新的对象，不改变原有的数据
midage_new.head(10)
```

```python
midage_new.loc[[99],['Pclass','Sex']] 
```

### 1.5.4 任务四：使用loc方法将midage的数据中第100，105，108行的"Pclass"，"Name"和"Sex"的数据显示出来

```python
midage_new.loc[[100,105,108],['Pclass','Name','Sex']] # loc, lable based location
```

### 1.5.5 任务五：使用iloc方法将midage的数据中第100，105，108行的"Pclass"，"Name"和"Sex"的数据显示出来

```python
midage_new.iloc[[100,105,108],[2,3,4]] # iloc, integer based location
```

### 1.6.1 任务一：利用Pandas对示例数据进行排序，要求升序

排序sorting分两种情况：1.根据DataFrame各个轴的索引排序 2.根据series的值进行排序

```python
frame = pd.DataFrame(np.arange(8).reshape((2,4)),index=['2','1'],columns=['d','a','b','c'])
frame 
```

【问题】：大多数时候我们都是想根据列的值来排序,所以将你构建的DataFrame中的数据根据某一列，升序排列

```python
frame.sort_values(by='d',ascending=True) #根据series的值进行牌型
```

【总结】下面将不同的排序方式做一个总结

```python
# 1. 行索引，升序排序
frame.sort_index(axis=0,ascending=True)
```

```python
# 2.  列索引，升序排序
frame.sort_index(axis=1,ascending=True)
```

```python
# 3. 行索引，降序排序
frame.sort_index(axis=0,ascending=False)
```

```python
# 4. 多列数据，降序排序
frame.sort_values(by=['d','a'],ascending=False)
```

### 1.6.3 任务三：利用Pandas进行算术计算，计算两个DataFrame数据相加结果

```python
frame_a = pd.DataFrame(np.arange(9.).reshape(3, 3),
                     columns=['a', 'b', 'c'],
                     index=['one', 'two', 'three'])
frame_b = pd.DataFrame(np.arange(12.).reshape(4, 3),
                     columns=['a', 'e', 'c'],
                     index=['first', 'one', 'two', 'second'])
frame_a
```

```python
frame_b
```

```python
# dataframe间进行加法运算
frame_a + frame_b  # 直接利用数学符号进行算数运算
```

```python
# 两个DataFrame对象间进行算数运算时，如果运算后的结果，某些项在原先的两个DataFrame对象中均不存在时，如何进行填充呢
frame_a.add(frame_b,fill_value=0) 
# 可以看到，df1.add(df2,fill_value=0)方法也不是万能的，
# 当运算后的结果，某些项在df1和df2中均不存在时，无法被填充，依然是nan
```
